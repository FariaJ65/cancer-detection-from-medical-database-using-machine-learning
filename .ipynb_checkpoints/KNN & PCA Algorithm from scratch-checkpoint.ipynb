```python
# STEP 1: substract the mean from feature
x_mean=feature-np.mean(feature,axis=0)
```

```python
x_mean.values
```

```python
#STEP 2: calculate covariance matrix
cov_matrix=np.cov(x_mean,rowvar=False) 
```

```python
cov_matrix
```

```python
#STEP 3: claculate eigenvalues and eigenvectors of the covariance matrix
# linalg.eigh() method returns the eigenvalues and eigenvectors of matrix
eigan_value,eigan_vector=np.linalg.eigh(cov_matrix)
```

```python
print(eigan_value)
```

```python
print(eigan_vector)
```

```python
print(np.shape(eigan_vector),np.shape(eigan_value))
```

```python
#Step 4: Sorting eigan values in decending order
sorted_index=np.argsort(eigan_value)[::-1]
```

```python
print(np.argsort(eigan_value))
```

```python
sorted_index
```

```python
sorted_eiganValue=eigan_value[sorted_index]
```

```python
sorted_eiganValue
```

```python
#Similarly Sorting eigan vector
sorted_eiganVector=eigan_vector[:,sorted_index]
```

```python
sorted_eiganVector
```

```python
#STEP 5:selecting the first 2 eigenvectors of our final reduced dataset
eigenVector_subset=sorted_eiganVector[:,0:2]
```

```python
eigenVector_subset
```

```python
#STEP 6:transform the data
#Originally it was 97*120, now its 97*2
xReduced=(np.dot(eigenVector_subset.transpose(),x_mean.transpose())).transpose()
```

```python
xReduced
```

```python
np.shape(xReduced)
```

```python
#Creating a Pandas DataFrame of reduced Dataset
principalDf=pd.DataFrame(xReduced,columns=['PC1','PC2'])
```

```python
#combine both to create complete dataset
principalDf=pd.concat([principalDf,res],axis=1)
```

```python
def euclidean_distance(x1, x2):
        return np.sqrt(np.sum((x1 - x2)**2))


class KNN:

    def __init__(self, k=5):
        self.k = k

    def fit(self, xt, yt):
        self.xtrain = xt
        self.ytrain = yt

    def predict(self, xt):
        y_pred = [self.classify(x) for x in xt]
        return np.array(y_pred)

    def classify(self, x):
        # Compute distances between x and all examples in the training set
        distances = [euclidean_distance(x, xtrain) for xtrain in self.xtrain]
        print(distances)
        print("\n")
        # Sort by distance and return indices of the first k neighbors
        k_idx = np.argsort(distances)[:self.k]
        print( np.sort(distances)[:self.k])
        print("\n")
        print(k_idx)
        print("\n")
        # Extract the labels of the k nearest neighbor training samples
        k_neighbor_labels = [self.ytrain[i] for i in k_idx]  
        print(k_neighbor_labels)
        print("\n")
        # return the most common class label
        most_common = Counter(k_neighbor_labels).most_common(1)
        print(most_common)
        print("\n")
        return most_common[0][0]
```

```python
kn=5
clf = KNN(k=kn)
clf.fit(xtrain, ytrain)
predictions = clf.predict(xtest)
```

```python
predictions.shape
```

```python
#print(classification_report(ytest, predictions))
print(confusion_matrix(ytest, predictions))
```
