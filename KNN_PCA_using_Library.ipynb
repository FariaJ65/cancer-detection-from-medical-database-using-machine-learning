```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Abnormality.csv")
```

```python
dataset
```

```python
#data
x=dataset.iloc[:,1:119]
```

```python
x.head()
```

```python
y=dataset.iloc[:,-1]
```

```python
y.tail()
```

```python
# Split dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.40) 
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier
#Create KNN Classifier
knn = KNeighborsClassifier()
```

```python
#Train the model using the training sets
knn.fit(x_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = knn.predict(x_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
x_pca = pca.fit(x).transform(x)
```

```python
x_pca
```

```python
Df_pca=pd.DataFrame(x_pca,columns=['PC1','PC2','PC3'])
```

```python
res=pd.DataFrame(y,columns=['class'])
```

```python
Df_pca=pd.concat([Df_pca,res],axis=1)
```

```python
Df_pca
```

```python
x2=Df_pca.iloc[:,0:3]
```

```python
x2
```

```python
y2=Df_pca.iloc[:,-1]
```

```python
y2
```

```python
Xax = Df_pca.iloc[:,0:1]
Yax = Df_pca.iloc[:,0:2]
Zax = Df_pca.iloc[:,0:3]
```

```python
cdict = {0:'red',1:'green'}
labl = {0:'Malignant',1:'Benign'}
marker = {0:'*',1:'o'}
alpha = {0:.3, 1:.5}
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')

fig.patch.set_facecolor('white')
for l in np.unique(y):
    ix=np.where(y==l)
    ax.scatter(Xax[ix], Yax[ix], Zax[ix], c=cdict[l], s=40,label=labl[l], marker=marker[l], alpha=alpha[l])
# for loop ends
ax.set_xlabel("First Principal Component", fontsize=14)
ax.set_ylabel("Second Principal Component", fontsize=14)
ax.set_zlabel("Third Principal Component", fontsize=14)

ax.legend()
plt.show()
```

```python
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
plt.rcParams['legend.fontsize']=10
ax.plot(Xax,Yax,Zax,'o',markersize=8,alpha=0.5,color='blue',label="CLASS 0")
ax.plot(Xax,Yax,Zax,'o',markersize=8,alpha=0.5,color='red',label="CLASS 1")
ax.set_xlabel(' PC1')
ax.set_ylabel(' PC2')
ax.set_zlabel(' PC3')
plt.title('3D projection')
ax.legend(loc='upper right')
plt.savefig('3Dfigure.png') 
plt.show()
```
